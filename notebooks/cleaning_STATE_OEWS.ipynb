```python
import pandas as pd
import re
```

```python
st2017 = pd.read_csv('../data/OEWS/2017state_M_dl.csv')
st2017.head(3)
```

```python
st2017 = st2017[['AREA', 'ST', 'STATE','OCC_CODE', 'OCC_TITLE', 'TOT_EMP', 'JOBS_1000', 'LOC_Q', 'A_MEAN', 
                         'A_PCT10', 'A_PCT25', 'A_MEDIAN', 'A_PCT75', 'A_PCT90']]
st2017.loc[:, 'year'] = 2017
st2017.head(2)
```

```python
col2017 = pd.Series(['AREA', 'ST', 'STATE','OCC_CODE', 'OCC_TITLE', 'TOT_EMP', 'JOBS_1000', 'LOC_Q', 'A_MEAN', 
                         'A_PCT10', 'A_PCT25', 'A_MEDIAN', 'A_PCT75', 'A_PCT90', 'year'])
lowcol2017 = col2017.str.lower()
st2017.columns = lowcol2017
st2017.head(3)
```

```python
bad1 = ['*', '**', '#']
good1 = [0, 0, 145601]
st2017 = st2017.replace(to_replace=bad1, value=good1)
st2017.head(3)
```

```python
occ_codes_2017 = ['11-3021', '13-1081', '13-1111', '13-1161', '15-1111', '15-1121', '15-1122', '15-1131', '15-1132', '15-1133',
                '15-1134', '15-1141', '15-1142', '15-1143', '15-1151', '15-1152', '15-1199', '15-2031', '15-2041', '17-2061',
                '43-9011', '43-9021', '43-9111', '49-2011']
st2017 = st2017.loc[st2017.occ_code.isin(occ_codes_2017)].reset_index(drop=True)

print(st2017.shape)
st2017.tail()
```

```python
notstates2017 = ['Guam', 'Virgin Islands']
st2017 = st2017.loc[~st2017.state.isin(notstates2017)].reset_index(drop=True)
st2017.tail()
```
