### Append the dataframes

```python
commsa2000 = msa2000_1.append(msa2000_2, ignore_index=True, verify_integrity=True, sort=False)
commsa2000.info()
```

### Keep only necessary columns & rename columns to match 2010/2017 column names.

```python
commsa2000 = commsa2000[['prim_state', 'area', 'area_name', 'occ_code', 'occ_titl', 'tot_emp', 'a_mean', 'a_wpct10', 
                         'a_wpct25', 'a_median', 'a_wpct75', 'a_wpct90', 'year']]

commsa2000 = commsa2000.rename(columns= {'occ_titl': 'occ_title','a_wpct10':'a_pct10', 'a_wpct25' : 'a_pct25', 
                                         'a_wpct75': 'a_pct75', 'a_wpct90': 'a_pct90'})
commsa2000.head(2)
```

### Keep only relevant OCC codes for the year 2000.

```python
occ_codes_2000 = ['11-3021', '13-1111', '15-1011', '15-1021', '15-1031',                                                      
                       '15-1032', '15-1041', '15-1051', '15-1061', '15-1071', 
                       '15-1081', '15-2031', '15-2041', '17-2061','19-3021',
                       '43-9011', '43-9021', '43-9111', '49-2011']
commsa2000 = commsa2000.loc[commsa2000.occ_code.isin(occ_codes_2000)].reset_index(drop=True)
commsa2000
```

### Clean up unhelpful values in the data set.

(* = indicates that a wage estimate is not available" - replace with 0)

(** = indicates that an employment estimate is not available - replace with 0)

(# = indicates a wage greater than $70.00 per hour or $145,600 per year" - replace with $145,601)

```python
bad = ['*', '**', '#']
good = [0, 0, 145601]
```

```python
commsa2000 = commsa2000.replace(to_replace=bad, value=good)
```

```python
commsa2000
```

### Clean out "MSA" from area_name column. Need values that Tableau can use to create a map.

```python
commsa2000['area_name'] = commsa2000['area_name'].str.replace("MSA", "")
```

```python
assert commsa2000['area_name'].str.contains('MSA').any() == False
```

```python
commsa2000.head()
```

```python
import re
```

### Use a regular expression and str.replace to clean the space P off the end of some area_name values

```python
commsa2000['area_name'] = commsa2000['area_name'].str.replace(r' P$', '', regex=True)
print (commsa2000)
```
